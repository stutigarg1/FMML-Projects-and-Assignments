```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
# Removing rows which have missing values

data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
# Dividing the dataset into features (X) and the target variable (y)

X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data
np.random.seed(0)
X = pd.DataFrame({
    'make': np.random.choice(['Toyota', 'Honda', 'Ford', 'BMW'], size=100)
})
y = np.random.normal(loc=20000, scale=5000, size=100)

# Create a figure and a grid of subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Swarm plot on the first subplot
sns.swarmplot(x=X["make"], y=y, ax=ax1)
ax1.set_title('Car Maker vs Price - Swarm Plot')

# Violin plot on the second subplot
sns.violinplot(x=X["make"], y=y, palette="Set3", ax=ax2)
ax2.set_title('Car Maker vs Price - Violin Plot')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
